#### Question 1. From the 'advertiser_stats' table, list the top 20 advertisers and sort them by their total spending in USD 


```
# your code here
import pandas as pd

df1 = pd.io.gbq.read_gbq('''
  SELECT advertiser_id, SUM(spend_usd) as Q1_total_spending   # USE advertiser_id. NOT advertiser_name.
  FROM `bigquery-public-data.google_political_ads.advertiser_stats`
  GROUP BY advertiser_id
  ORDER BY Q1_total_spending DESC
  LIMIT 20
''', project_id=project_id, dialect='standard')

df1.head(20)
```

#### Question 2. From the 'geo spend' table, list the top 10 states by their total spending in USD

```
# your code here
df2 = pd.io.gbq.read_gbq('''
  SELECT country_subdivision_primary, SUM(spend_usd) as Q2_total_spending # Use US states. NOT country
  FROM `bigquery-public-data.google_political_ads.geo_spend`
  GROUP BY country_subdivision_primary
  ORDER BY Q2_total_spending DESC
  LIMIT 10
''', project_id=project_id, dialect='standard')

df2.head(10)
```

#### Question 3. From the 'creative_stats' table, which advertiser had their ads running for the most number of days during the years 2018 and 2019

```
# your code here
df3 = pd.io.gbq.read_gbq('''
  SELECT advertiser_name,
  SUM(DATE_DIFF(
     IF(CAST(date_range_end AS DATE) <= CAST('2019-12-31' AS DATE), CAST(date_range_end AS DATE), CAST('2019-12-31' AS DATE)), 
     IF(CAST(date_range_start AS DATE) >= CAST('2018-01-01' AS DATE), CAST(date_range_start AS DATE), CAST('2018-01-01' AS DATE)),
     DAY)
    ) AS Q3_the_most_days, 
  FROM `bigquery-public-data.google_political_ads.creative_stats`
  WHERE (CAST(date_range_start AS DATE) <= CAST('2019-12-31' AS DATE)) AND (CAST(date_range_end AS DATE) >= CAST('2018-01-01' AS DATE))
  GROUP BY advertiser_name
  ORDER BY Q3_the_most_days DESC
  LIMIT 1
''', project_id=project_id, dialect='standard')
df3.head()
```

Question 4. From the 'covid19_usadacts.deaths' table, list the top 5 counties with the most death cases in 2020

```
df4 = pd.io.gbq.read_gbq('''
  SELECT * 
  FROM `bigquery-public-data.covid19_usafacts.deaths`
 ''', project_id=project_id, dialect='standard')

column_filtered = []
for column in df4.columns:
  year = column[1:5]
  if year == '2020':
    column_filtered.append(column)

df4_1 = pd.DataFrame([])
df4_1['county_name'] = df4['county_name']
df4_1['Q4_the_most_death_cases'] = df4[column_filtered].sum(axis=1)
df4_1.sort_values('Q4_the_most_death_cases', ascending=False).head()
```

#### Question 5. From the advertiser_stats, list the heaviest spending advertiser by calculating total_spending divided by total_creatives. For this you need to divide the total_spend_usd / total_creatives to get a ratio of their spending per ad. Sort the list and show only the top 10. 

```
# your code here
df5_1 = pd.io.gbq.read_gbq('''
  SELECT advertiser_id, SUM(spend_usd) AS total_spend_usd
  FROM `bigquery-public-data.google_political_ads.advertiser_stats`
  GROUP BY advertiser_id       # USE advertiser_id NOT advertiser_name
''', project_id=project_id, dialect='standard').set_index('advertiser_id')

df5_2 = pd.io.gbq.read_gbq('''
  SELECT advertiser_id, total_creatives
  FROM `bigquery-public-data.google_political_ads.advertiser_stats`
''', project_id=project_id, dialect='standard').set_index('advertiser_id')

df5 = df5_1.join(df5_2, on='advertiser_id')
df5['Q5_the_heaviest_spending'] = 1. * df5['total_spend_usd'] / df5['total_creatives']
df5.sort_values('Q5_the_heaviest_spending', ascending=False).head(10)
```
